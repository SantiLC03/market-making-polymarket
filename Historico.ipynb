# Análisis del histórico de mercado

### Importamos Datos

```python
import pandas as pd
data = pd.read_csv('Data/Real Madrid 2025 Polymarket.csv')
print(data.head(6))
```

```python
import matplotlib.pyplot as plt
buys = data[data['SIDE'] == 'buy']
sells = data[data['SIDE'] == 'sell']

plt.scatter(buys['TIMESTAMP'], buys['PRICE'], color='green', label='Buys')
plt.xlabel('Timestamp')    
```
